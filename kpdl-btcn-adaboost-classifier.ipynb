- Một mô hình phân loại Mi được học cho mỗi tập huấn luyện Di

# **3. Phân lớp AdaBoost** <a class="anchor" id="3"></a>

[Back to Notebook Contents](#0.1)


- **AdaBoost hay Adaptive Boosting** là một trong những công cụ phân loại tăng cường tổng hợp được đề xuất bởi Yoav Freund và Robert Schapire vào năm 1996.

- Nó kết hợp nhiều bộ phân loại yếu để tăng độ chính xác của bộ phân loại.

- AdaBoost là một phương thức tổng hợp lặp đi lặp lại. Trình phân loại AdaBoost xây dựng một trình phân loại mạnh bằng cách kết hợp nhiều trình phân loại hoạt động kém để bạn có được trình phân loại mạnh có độ chính xác cao.

- Khái niệm cơ bản đằng sau Adaboost là thiết lập trọng số của bộ phân loại và đào tạo mẫu dữ liệu trong mỗi lần lặp để đảm bảo dự đoán chính xác về các quan sát bất thường.

- Bất kỳ thuật toán học máy nào cũng có thể được sử dụng làm bộ phân loại cơ sở nếu nó chấp nhận các trọng số trên tập huấn luyện.


- **AdaBoost** cần đạt 2 điều kiện:

   1. Phân loại nên được đào tạo tương tác trên các ví dụ đào tạo được cân nhắc khác nhau.
  
   2. Trong mỗi lần lặp, nó cố gắng cung cấp sự phù hợp tuyệt vời cho các ví dụ này bằng cách giảm thiểu lỗi đào tạo.

- Để xây dựng bộ phân loại AdaBoost, hãy tưởng tượng rằng với tư cách là bộ phân loại cơ sở đầu tiên, chúng tôi đào tạo thuật toán Cây quyết định để đưa ra dự đoán về dữ liệu đào tạo của chúng tôi.

- Bây giờ, theo phương pháp của AdaBoost, trọng lượng của các phiên bản đào tạo được phân loại sai được tăng lên.

- Bộ phân loại thứ hai được huấn luyện và ghi nhận các trọng số được cập nhật và nó lặp đi lặp lại quy trình.

- Vào cuối mỗi mô hình dự đoán, chúng ta kết thúc việc tăng trọng số của các cá thể được phân loại sai để mô hình tiếp theo thực hiện tốt hơn chúng, v.v.

- AdaBoost thêm các yếu tố dự đoán vào nhóm dần dần làm cho nó trở nên tốt hơn. Nhược điểm lớn của thuật toán này là mô hình không thể được song song hóa vì mỗi dự đoán chỉ có thể được đào tạo sau khi người trước đó đã được đào tạo và đánh giá.

- Dưới đây là các bước để thực hiện thuật toán AdaBoost:

  1. Ban đầu, tất cả các quan sát được cho có trọng lượng bằng nhau.

  2. Một mô hình được xây dựng trên một tập hợp con dữ liệu.

  3. Sử dụng mô hình này, các dự đoán được thực hiện trên toàn bộ tập dữ liệu.

  4. Sai số được tính toán bằng cách so sánh các dự đoán và giá trị thực tế.

  5. Khi tạo mô hình tiếp theo, trọng số cao hơn được đưa ra cho các điểm dữ liệu đã được dự đoán không chính xác.

  6. Trọng lượng có thể được xác định bằng cách sử dụng giá trị lỗi. Ví dụ, sai số càng cao thì trọng số được gán cho quan sát càng nhiều.
  
  7. Quá trình này được lặp lại cho đến khi hàm lỗi không thay đổi hoặc đạt đến giới hạn tối đa của số lượng bộ ước lượng.

### 6.5 Xây dựng model AdaBoost <a class="anchor" id="6.5"></a>

```python
# Nhập về thư viện phân lớp AdaBoost từ sklearn
from sklearn.ensemble import AdaBoostClassifier

# Tạo đối tượng phân lớp AdaBoost với hàm của sklearn
adb_c = AdaBoostClassifier(n_estimators = 15, learning_rate = 1, random_state = 0)
```

```python
# Hàm tự định nghĩa thuật toán AdaBoost

# Nhập về thư viện phân lớp Cây quyết định của sklearn sử dụng làm người học cơ sở
from sklearn.tree import DecisionTreeClassifier

# Hàm kiểm tra giá trị của y
def check(y):
    assert set(y) == {-1,1}
    return y

# Hàm tạo model AdaBoost
def init_model(iters, X):
    n = X.shape[0]
    sample_weights = np.zeros((iters, n))
    stumps = np.zeros(iters, dtype= object)
    stump_weights = np.zeros(iters)
    errors = np.zeros(iters)
    return stumps, stump_weights, sample_weights, errors

# Hàm đồ họa minh họa thuật toán: biểu diễn các điểm và decision boundary
def plot_adaboost(X: np.ndarray,
                  y: np.ndarray,
                  stumps= None, stump_weights= None, roll = 0,
                  clf= None,
                  sample_weights: Optional[np.ndarray] = None,
                  ax: Optional[mpl.axes.Axes] = None):

    y = check(y) # Kì vọng nhãn bằng ±1

    if not ax:
        fig, ax = plt.subplots(figsize=(5, 5), dpi=100)
        fig.set_facecolor('white')

    pad = 1
    x_min, x_max = X[:, 0].min() - pad, X[:, 0].max() + pad
    y_min, y_max = X[:, 1].min() - pad, X[:, 1].max() + pad

    if sample_weights is not None:
        sizes = np.array(sample_weights) * X.shape[0] * 100  
    else:
        sizes = np.ones(shape=X.shape[0]) * 100

    X_pos = X[y == 1]
    sizes_pos = sizes[y == 1]
    ax.scatter(*X_pos.T, s=sizes_pos, marker='+', color='red')

    X_neg = X[y == -1]
    sizes_neg = sizes[y == -1]
    ax.scatter(*X_neg.T, s=sizes_neg, marker='.', c='blue')

    if clf:
        plot_step = 0.01
        xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                             np.arange(y_min, y_max, plot_step))
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)

        # If all predictions are positive class, adjust color map acordingly
        if list(np.unique(Z)) == [1]:
            fill_colors = ['r']
        else:
            fill_colors = ['b', 'r']

        ax.contourf(xx, yy, Z, colors=fill_colors, alpha=0.2)

    if roll:
        plot_step = 0.01
        xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                             np.arange(y_min, y_max, plot_step))

        Z = predict(np.c_[xx.ravel(), yy.ravel()], stumps, stump_weights)
        Z = Z.reshape(xx.shape)

        # If all predictions are positive class, adjust color map acordingly
        if list(np.unique(Z)) == [1]:
            fill_colors = ['r']
        else:
            fill_colors = ['b', 'r']

        ax.contourf(xx, yy, Z, colors=fill_colors, alpha=0.2)

    ax.set_xlim(x_min+0.5, x_max-0.5)
    ax.set_ylim(y_min+0.5, y_max-0.5)
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    
# Hàm triển khai mô hình Adaboost
def AdaBoostClf(X, y, iters= 10):
    n = X.shape[0]
    # Check y
    y = check(y)
    # Initialize
    stumps, stump_weights, sample_weights, errors = init_model(iters= iters, X= X)

    # First weight = 1/n
    sample_weights[0] = np.ones(shape= n) / n

    for i in range(iters):
        # Fit for stump: weak learner
        current_sew = sample_weights[i]
        stump = DecisionTreeClassifier(max_depth= 1, max_leaf_nodes= 2)
        stump = stump.fit(X, y, sample_weight= current_sew)

        # Calculate error
        stump_pred = stump.predict(X)
        error = current_sew[stump_pred != y].sum()
        stump_weight = np.log((1 - error) / error) / 2

        # New sample weight
        new_sew = current_sew * np.exp(-1 * stump_weight * y * stump_pred)

        # Renormalize weights
        new_sew = new_sew / new_sew.sum()

        # If not last iter, update sample weights for i+1
        if (i + 1) < iters:
            sample_weights[i+1] = new_sew

        # Save result
        errors[i] = error
        stumps[i] = stump
        stump_weights[i] = stump_weight

    return stumps, stump_weights, sample_weights

# Hàm dự đoán dựa trên mô hình
def predict(X, stumps, stump_weights):
    stump_preds = np.array([stump.predict(X) for stump in stumps])
    return np.sign(np.dot(stump_weights, stump_preds))
```

### Tạo phân loại Adaboost

- Các tham số quan trọng nhất là `base_estimator`, `n_estimators` và `learning_rate`.

- **base_estimator** là thuật toán học tập được sử dụng để đào tạo các mô hình yếu. Giá trị này hầu như không cần phải thay đổi vì cho đến nay giá trị học phổ biến nhất sử dụng khi AdaBoost là cây quyết định - đối số mặc định của tham số này.

- **n_estimators** là số lượng mô hình để đào tạo lặp đi lặp lại.

- **learning_rate** là đóng góp của mỗi mô hình vào trọng số và mặc định là 1. Giảm tốc độ học sẽ có nghĩa là trọng số sẽ được tăng hoặc giảm ở một mức độ nhỏ, nó khiến việc đào tạo mô hình diễn ra chậm hơn (nhưng đôi khi dẫn đến điểm hiệu suất tốt hơn).

### 6.7 Đồ họa hóa quá trình thực hiện<a class="anchor" id="6.7"></a>


```python
# Lập các hàm vẽ biểu đồ

def truncate_adaboost(stumps, stump_weights, t: int):
    assert t > 0, 't must be a positive integer'
    x = stumps[:t]
    y = stump_weights[:t]
    return x, y

def plot_staged_adaboost(X, y, stumps, sample_weights, stump_weights, start=0, end=10):
    # larger grid
    fig, axes = plt.subplots(figsize=(8, (end-start)*3),
                             nrows=(end-start),
                             ncols=2,
                             sharex=True,
                             dpi=100)

    fig.set_facecolor('white')
    for i in range(start, end):
        ax1, ax2 = axes[i-start]

        # Plot weak learner
        ehe = ax1.set_title(f'Weak learner at t={i + 1}')
        plot_adaboost(X, y, clf= stumps[i],
                      sample_weights= sample_weights[i],
                      annotate=False, ax=ax1)

        # Plot strong learner
        new_stumps,  new_sw = truncate_adaboost(stumps, stump_weights, t=i + 1)
        ehe = ax2.set_title(f'Strong learner at t={i + 1}')
        plot_adaboost(X, y, new_stumps, new_sw, sample_weights= sample_weights[i], roll=1, annotate=False, ax=ax2)

    plt.tight_layout()
    plt.subplots_adjust(top=0.95)
    plt.show()
```

```python
# Áp dụng vào model từ cây thứ 10
plot_staged_adaboost(X1_test, y1_test, stumps, sample_weights, stump_weights, start=9, end=15)
```
